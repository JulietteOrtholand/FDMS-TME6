```python
%load_ext autoreload
%autoreload 2

import pandas as pd
import matplotlib.pyplot as plt 
import numpy as np

from TME6 import * 
```

# Données

```python
data = pd.read_csv('CTR.txt',sep = ';|:', header=None)
```

# Baselines

```python
randCumul = random_bandit(data)
plt.plot(range(len(randCumul)),randCumul)
plt.title('Random Bandit')
plt.show()
```

```python
statCumul = staticBest(data)
plt.plot(range(len(statCumul)), statCumul)
plt.title('Statitic Best Bandit')
plt.show()
```

```python
optimCumul = optimale(data)
plt.plot(range(len(optimCumul)), optimCumul)
plt.title('Bandit Optimal')
plt.show()
```

# Résumé Baseline

```python
plt.plot(range(len(randCumul)),randCumul, label = 'Random Bandit')
plt.plot(range(len(statCumul)),statCumul, label = 'Statitic Best Bandit')
plt.plot(range(len(optimCumul)),optimCumul, label = 'Bandit Optimal')
plt.title('Résumé')
plt.legend()
plt.show()
```

# UCB

```python
ucbCumul = ucb(data)
plt.plot(range(len(ucbCumul)), ucbCumul)
plt.title('Bandit UCB')
plt.show()
```

```python
plt.plot(range(len(randCumul)),randCumul, label = 'Random Bandit')
plt.plot(range(len(statCumul)),statCumul, label = 'Statitic Best Bandit')
plt.plot(range(len(optimCumul)),optimCumul, label = 'Bandit Optimal')
plt.plot(range(len(ucbCumul)),ucbCumul, label = 'Bandit UCB')
plt.title('Résumé')
plt.legend()
plt.show()
```

# LinUCB

```python
list_alpha = [0.1,0.5,1.,1.5,2]
res_linubc = []
for alpha in list_alpha:
    linucbCumul = linucb(data,alpha)
    plt.plot(range(len(linucbCumul)), linucbCumul,label = 'alpha = '+ str(alpha))
    res_linubc.append(linucbCumul)
plt.title('Bandit linUCB')
plt.legend()
plt.show()
```

```python
plt.plot(range(len(randCumul)),randCumul, label = 'Random Bandit')
plt.plot(range(len(statCumul)),statCumul, label = 'Statitic Best Bandit')
plt.plot(range(len(optimCumul)),optimCumul, label = 'Bandit Optimal')
plt.plot(range(len(ucbCumul)),ucbCumul, label = 'Bandit UCB')
plt.plot(range(len(res_linubc[0])),res_linubc[0], label = 'Bandit linUCB')
plt.title('Résumé')
plt.legend()
plt.show()
```
